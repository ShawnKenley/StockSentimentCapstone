```python
investa_df = pd.DataFrame([], columns = ['Date_Time', 'Stock_Code', 'Headline', 'Summary', 'Links'])

headers = soup.find_all('td')
count = (950+1)*(9*2)

for head in headers[1:count:2]:
    dt, stock_code, headline, summary, link = np.nan, np.nan, np.nan, np.nan, np.nan
    try:
        tag1=head.find('h4',class_="media-heading")
        stock_code = tag1.find('a').get_text()
        dt = tag1.find('small').get_text()
        
    except:
        pass
    
    try:
        tag2=head.find('p',class_="text-break-word")
        headline = tag2.find('strong').get_text()
        summary = tag2.find('span').get_text()
        
    except:
        pass
    
    try:        
        tag3=head.find('li',class_="media-post-action-individual")
        link = tag3.find('a').get('href')

        
        data = pd.DataFrame([[dt, stock_code, headline, summary, link]], 
                            columns = ['Date_Time', 'Stock_Code', 'Headline', 'Summary', 'Links'])
        investa_df = investa_df.append(data)        
        
    except:
        data = pd.DataFrame([[dt, stock_code, headline, summary, link]], 
                            columns = ['Date_Time', 'Stock_Code', 'Headline', 'Summary', 'Links'])
        investa_df = investa_df.append(data) 
```

```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
fil_abs_df = pd.DataFrame([], columns = ['Links', 'Fil_bodies', 'En_bodies'])


for i in range(len(abs_links)):
    url = abs_links[i]
    website_url = requests.get(url,
                              headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64)'}
                              ).text

    soup = BeautifulSoup(website_url, "html.parser")
    
    headers = soup.find_all('div', class_="article-content")
    headers += (soup.find_all('small', class_="media-caption"))

    for head in headers:
        print('link number:', i)
        fil_captions = ''
        tags = head.find_all('p')
        for caption in tags:
            fil_captions = fil_captions + ' ' + caption.get_text()
            #print(caption.get_text())
        
        
        data = pd.DataFrame([[abs_links[i], fil_captions, np.nan]], 
                            columns = ['Links', 'Fil_bodies', 'En_bodies'])
        fil_abs_df = fil_abs_df.append(data)
```

```python
fil_abs_df.reset_index().drop(columns = 'index').to_csv('fil_abs_df.csv', index=False)
```
